### Importing all the datasets

```python
import pandas as pd

ldap_data = pd.read_csv('../data/core/LDAP.csv', low_memory=False)
mssql_data = pd.read_csv('../data/core/MSSQL.csv', low_memory=False)
netbios_data = pd.read_csv('../data/core/NetBIOS.csv', low_memory=False)
portmap_data = pd.read_csv('../data/core/Portmap.csv', low_memory=False)
syn_data = pd.read_csv('../data/core/SYN.csv', low_memory=False)
udp_data = pd.read_csv('../data/core/UDP.csv', low_memory=False)
udplag_data = pd.read_csv('../data/core/UDPLag.csv', low_memory=False)
```

##### Comparing the datasets

```python
# 1. Put all your dataframes into a dictionary with their names
dataframes = {
    'LDAP': ldap_data,
    'MSSQL': mssql_data,
    'NetBIOS': netbios_data,
    'Portmap': portmap_data,
    'SYN': syn_data,
    'UDP': udp_data,
    'UDPLag': udplag_data
}

# 2. Create a new DataFrame from the .columns attribute of each one
#    We use pd.Series because it handles lists of different lengths gracefully
comparison_df = pd.DataFrame({
    name: pd.Series(df.columns) for name, df in dataframes.items()
})

# 3. Display the result
#    This will show you all column names side-by-side
print("Side-by-side column comparison:")
display(comparison_df)

# --- Optional: Check for any differences ---
# This will show you exactly which row (column index) has a mismatch
print("\nRows with *any* differences:")
# We check if, for any row, not all values are the same as the first value
row_has_difference = comparison_df.apply(lambda row: not all(row == row[0]), axis=1)
display(comparison_df[row_has_difference])
```
